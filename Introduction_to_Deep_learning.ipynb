b. **Query function**

We will need to transport the signals from the input nodes to the nodes of the hidden layer. This denotes the signals going to the hidden layer. For this, we will need a matrix multiplication: 
$X_{input, hidden}  = Weight_{input, hidden}* Input$.



The signal coming out from the hidden layer is given by
$O_{hidden} = sigmoid(X_{input, hidden})$

We will also need to transport the signals from the hidden nodes of the output layer  layer. This signal is going to the output layer. For this, we will need a matrix multiplication: 
$X_{hidden, output}  = Weight_{output, hidden}* O_{hidden}$.



The signal coming out of the output layer is given by
$O_{output,layer} = sigmoid(X_{hidden, output})$

```python
import scipy as sp
from scipy import signal
from scipy import special
```

```python
class my_neural_network:

  def __init__(self, input_nodes, hidden_nodes, out_put_nodes, learning_rate):

    self.input_nodes = input_nodes
    self.hidden_nodes = hidden_nodes
    self.out_put_nodes = out_put_nodes
    self.learning_rate = learning_rate
    
    # initalize the weights
    self.wweights_input_hidden = np.random.rand(self.hidden_nodes, self.input_nodes)- 0.5 # input layer to first hidden layer
    self.weights_hidden_ouput = np.random.rand(self.out_put_nodes, self.hidden_nodes) - 0.5 # input layer to first hidden layer

    # what about the activation function
    self.activation = lambda x: scipy.special.expit(x)

    pass
    
  def query(self, inputs_list):
    # convert inputs list to 2d array
    input_signal = np.array(inputs_list, ndmin=2).T
    signal_input_hidden = np.dot(self.wweights_input_hidden, input_signal)
    signal_out_of_hidden = self.activation(signal_input_hidden) # This is the signal that is coming out of the hidden layer
    # This signal will be going to the output layer. 

    signal_hidden_output = np.dot(self.wweights_hidden_ouput,signal_out_of_hidden)
    signal_out_of_output_layer = self.activation(signal_hidden_output)
  
    return signal_out_of_output_layer
```

c. **Training function**

This part of the construction of the nets contains 2 parts
1. Carry out a signal from the input layer to the output layer
2. Compare what comes of the layer and the real labels

The error is calculated by using
$$outputErrors = actualTarget - calculatedOutput$$.



One should notice that the error is coming from the fact that we randomly chose our weigths. Therefore, to minimize the errors from the output layer we need to go back and readjust the weights between the hidden layer and the output layer. Similarly, to adjust minimize the errors from the hidden layer we need to readjust the weights between the input layer and the hidden layer. For this, we are going back from output layer to the hidden layer.

$$errors_{hidden} = weights_{hidden,ouput}^T.OutputErrors$$.

To update the weight for the link between a node j and a node kâ€‹ in the next layer in matrix form:

$$Î”Ï‰_{jk} = Î± * E_{k} * sigmoid(O_{k}) * (1 - sigmoid(O_{k})). O^{T} _{j}$$


```python
class my_neural_network:

  def __init__(self, input_nodes, hidden_nodes, out_put_nodes, learning_rate):

    self.input_nodes = input_nodes
    self.hidden_nodes = hidden_nodes
    self.out_put_nodes = out_put_nodes
    self.learning_rate = learning_rate
    
    # initalize the weights
    self.weights_input_hidden = np.random.rand(self.hidden_nodes, self.input_nodes)- 0.5 # input layer to first hidden layer
    self.weights_hidden_ouput = np.random.rand(self.out_put_nodes, self.hidden_nodes) - 0.5 # input layer to first hidden layer

    # what about the activation function
    self.activation = lambda x: scipy.special.expit(x)

    pass
    
  def query(self, inputs_list):
    # convert inputs list to 2d array
    input_signal = np.array(inputs_list, ndmin=2).T
    signal_input_hidden = np.dot(self.wweights_input_hidden, input_signal)
    signal_out_of_hidden = self.activation(signal_input_hidden) # This is the signal that is coming out of the hidden layer
    # This signal will be going to the output layer. 

    signal_hidden_output = np.dot(self.wweights_hidden_ouput,signal_out_of_hidden)
    signal_out_of_output_layer = self.activation(signal_hidden_output)
  
    return signal_out_of_output_layer


  def training(self, inputs_list, target_list):
    ''' This function will allow us to propagate forward the signals
    then compare the computed target values with the actual ones'''

    # The first part of the training is the same as the query part
    # convert inputs list to 2d array
    input_signal = np.array(inputs_list, ndmin=2).T
    #convert the outputs list to 2d array
    actual_target= np.array(target_list, ndim = 2).T
    signal_input_hidden = np.dot(self.wweights_input_hidden, input_signal)
    signal_out_of_hidden = self.activation(signal_input_hidden) # This is the signal that is coming out of the hidden layer
    # This signal will be going to the output layer. 

    signal_hidden_output = np.dot(self.wweights_hidden_ouput,signal_out_of_hidden)
    signal_out_of_output_layer = self.activation(signal_hidden_output)

    final_outputs = signal_out_of_output_layer

    # Calculate the error from the output layer

    output_errors = actual_target - final_outputs

    # hidden layer error is the output_errors, split by weights, recombined at hidden nodes
    hidden_errors = np.dot(self.weights_hidden_ouput.T, output_errors)

    # update the weights for the links between the hidden and output layers
    self.weights_hidden_ouput += self.lr * np.dot((output_errors * final_outputs * (1.0 - final_outputs)),np.transpose(signal_out_of_hidden))

    # update the weights for the links between the input and hidden layers
    self.weights_input_hidden += self.lr * np.dot((hidden_errors *signal_hidden_output * (1.0 - signal_hidden_output)), np.transpose(inputs_list))

    pass
```

**Forward Propagation Function**

```python
# This function allows us to move the signal from the input layer to the output layer
def forward_propagation(signal,input_hidden_weights, hidden_output_weights):
  # transport signal from input layer to hidden layer
  first_operation = np.dot(input_hidden_weights, signal)
  # signal leaving the hidden layer
  first_transformed_signal = sigmoid_function(first_operation)
  # signal entering the output layer
  second_operation = np.dot(hidden_output_weights,first_transformed_signal)
  # signal coming from the output layer
  second_transformed_signal = sigmoid_function(second_operation)

  return first_operation, first_transformed_signal, second_operation, second_transformed_signal
```

**Regularization**

Sometimes, mostly without known, we empower our neural network with more than enough capacity. More often than not, this is leads our neural nets to be able to perfectly learn (memorize) from our training data. As a consequence, the neural network fails to generalize. To avoid this undesirable situation, it is a common practice to decomplexify our neural network. 

Among many other techniques, regularization comes as one of the most widely techniques to regularize the network. In the process of regularization, a dropout rate is always given. For example, suppose that we would like to have a neural net that has 2 hidden layers. Suppose that we want to put a dropout layer between the two layers ( meaning right after the first hidden layer). In addition, let us consider that the dropout rate between hidden layer 1 and hidden layer 2 is dropOutRate = 0.45. 

Suppose that the first layer has 256 neurons. The dropout will random remove that fraction n of units from participating in the next layer. After a dropout of 0.35 is applied, it will remain (1 - 0.45)*256 ~ 140 units from layer 1 participate in layer 2

**Activation Function**

It is a function that will decide whether a neuron should be activated or not. Therefore, it bounds the net input.

In deep Learning, several different types of activation functions. The most common ones are

```python
# Step function
```

$$ f(x) = 1, if x \ge 0 $$
$$ f(x) = 0, if x < 0 $$

```python
# Softmax
```

$$ softmax(x_{i}) = \frac{e^{x_{i}}}{âˆ‘_{i=0}^{N}e^{x_{i}}}$$

```python
# Rectified Linear Unit
```

$$ f(x) = max(0,x)$$

```python
# Leaky ReLu
```

$$ f(x) = ax, if x < 0 $$
$$ f(x) = x, otherwise $$


```python
# Let us write the sigmoid function
import numpy as np
def sigmoid(list_of_values):

  numerator = list(map(lambda x: np.exp(x), list_of_values))
  denominator = np.sum(numerator)
  return numerator/denominator
```

```python
# model is a 3-layer MLP with ReLU and dropout after each layer
model = Sequential()
model.add(Dense(hidden_units, input_dim=input_size))
model.add(Activation('relu'))
model.add(Dropout(dropout))
model.add(Dense(hidden_units))
model.add(Activation('relu'))
model.add(Dropout(dropout))
model.add(Dense(num_labels))
```

```python
# this is the output for one-hot vector
model.add(Activation('softmax'))
model.summary()
plot_model(model, to_file='mlp-mnist.png', show_shapes=True)
```

```python
# loss function for one-hot vector
# use of adam optimizer
# accuracy is a good metric for classification tasks
model.compile(loss='categorical_crossentropy',
 optimizer='adam',
 metrics=['accuracy'])
```

```python
# train the network
model.fit(x_train, y_train, epochs=20, batch_size=batch_size)
# validate the model on test dataset to determine generalization
loss, acc = model.evaluate(x_test, y_test, batch_size=batch_size)
print("\nTest accuracy: %.1f%%" % (100.0 * acc))
```
